### 단일 변수 히스토그램

```python
# 데이터 타입별 시각화 함수
def bar_plot(col, data, hue=None):
    f, ax = plt.subplots(figsize=(10, 5))
    sns.countplot(x=col, hue=hue, data=df, alpha=0.5)
    plt.show()

def dist_plot(col, data):
    f, ax = plt.subplots(figsize=(10, 5))
    sns.distplot(df[col], ax=ax)
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
    plt.show()

def bar_plot_ci(col, data):
    f, ax = plt.subplots(figsize=(10, 5))
    sns.barplot(x=col, y='target', data=df)
    plt.show()
```

```python
for col in obj_cols:
    bar_plot(col, df)
```

```python
for col in nobj_cols:
    dist_plot(col, df)
```

## 성별/연령과 질병간 관계 파악
>연령대가 높아질수록 위험도 수준이 올라가는 여성과 달리, 40-50대 중년 남성의 >발병 위험도는 60-80대 장년층과 비슷한 수준으로 높다.

### 고혈압

```python
# 남성 연령대별 고혈압
tmp = dat[dat['성별']==0]
ct = pd.crosstab(tmp['나이'], tmp['고혈압_k'])
cols = ct.columns.tolist()
cols = [ '정상', '고혈압전단계','1단계고혈압', '2단계고혈압']
ct = ct[cols]

totals = []
for i in range(len(ct)):
    totals.append(sum(ct.iloc[i, :]))

cl = ct.columns.to_list()

color_list = ['MistyRose','Salmon','Crimson', 'darkred']

for i in range(len(cl)):
    ct[cl[i]] = ct[cl[i]] / totals

fig = ct.plot.bar(stacked=True, figsize=(10,6), color=color_list)
plt.xticks(fontsize=12, rotation=360)
plt.yticks(fontsize=12)
plt.xlabel('나이', fontsize=12)
plt.title('남성 연령대별 고혈압', fontsize=16)
plt.legend(fontsize='x-large')
plt.show()
```
