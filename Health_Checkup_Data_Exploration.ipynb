## Data Exploration
범주형 변수로 데이터 타입 수정

```python
# GFR
df['GFR'] = df['GFR'].astype('int')
df['GFR'] = df['GFR'].astype('object')

# 신장질환
df['신장질환'] = df['신장질환'].astype('int')
df['신장질환'] = df['신장질환'].astype('object')

# risk_factor
df['risk_factor'] = df['risk_factor'].astype('int')
df['risk_factor'] = df['risk_factor'].astype('object')

# 단백뇨
df['단백뇨'] = df['단백뇨'].astype('int')
df['단백뇨'] = df['단백뇨'].astype('object')

# 위험도_허리둘레
df['위험도_허리둘레'] = df['위험도_허리둘레'].astype('int')
df['위험도_허리둘레'] = df['위험도_허리둘레'].astype('object')

# 시도코드
df['시도코드'] = df['시도코드'].astype('int')
df['시도코드'] = df['시도코드'].astype('object')

# 청력(좌)
df['청력(좌)'] = df['청력(좌)'].astype('int')
df['청력(좌)'] = df['청력(좌)'].astype('object')

# 청력(우)
df['청력(우)'] = df['청력(우)'].astype('int')
df['청력(우)'] = df['청력(우)'].astype('object')

# 복부비만
df['복부비만'] = df['복부비만'].astype('int')
df['복부비만'] = df['복부비만'].astype('object')

# 빈혈
df['빈혈'] = df['빈혈'].astype('int')
df['빈혈'] = df['빈혈'].astype('object')

# 요단백
df['요단백'] = df['요단백'].astype('int')
df['요단백'] = df['요단백'].astype('object')

# 흡연상태
df['흡연상태'] = df['흡연상태'].astype('int')
df['흡연상태'] = df['흡연상태'].astype('object')

# 성별
df['성별'] = df['성별'].astype('int')
df['성별'] = df['성별'].astype('object')

# 구강검진_수검 obj
df['구강검진_수검'] = df['구강검진_수검'].astype('int')
df['구강검진_수검'] = df['구강검진_수검'].astype('object')

# 실명 obj
df['실명'] = df['실명'].astype('int')
df['실명'] = df['실명'].astype('object')

# 고지혈증 obj
df['고지혈증'] = df['고지혈증'].astype('int')
df['고지혈증'] = df['고지혈증'].astype('object')

# 고혈압 obj
df['고혈압'] = df['고혈압'].astype('int')
df['고혈압'] = df['고혈압'].astype('object')

# 총콜레스테롤_div obj
df['총콜레스테롤_div'] = df['총콜레스테롤_div'].astype('int')
df['총콜레스테롤_div'] = df['총콜레스테롤_div'].astype('object')

# 공복혈당
df['공복혈당'] = df['공복혈당'].astype('int')
df['공복혈당'] = df['공복혈당'].astype('object')

# 당뇨
df['당뇨'] = df['당뇨'].astype('int')
df['당뇨'] = df['당뇨'].astype('object')

# 당뇨_표적장기손상
df['당뇨_표적장기손상'] = df['당뇨_표적장기손상'].astype('int')
df['당뇨_표적장기손상'] = df['당뇨_표적장기손상'].astype('object')

# whtr
df['whtr'] = df['whtr'].astype('int')
df['whtr'] = df['whtr'].astype('object')

# bmi_div
df['bmi_div'] = df['bmi_div'].astype('int')
df['bmi_div'] = df['bmi_div'].astype('object')

# BMI
df['BMI'] = df['BMI'].astype('int')
df['BMI'] = df['BMI'].astype('object')

# 대사증후군
df['대사증후군'] = df['대사증후군'].astype('int')
df['대사증후군'] = df['대사증후군'].astype('object')

# 나이
df['나이'] = df['나이'].astype('int')
df['나이'] = df['나이'].astype('object')
```

### 단일 변수 히스토그램

```python
# 데이터 타입별 시각화 함수
def bar_plot(col, data, hue=None):
    f, ax = plt.subplots(figsize=(10, 5))
    sns.countplot(x=col, hue=hue, data=df, alpha=0.5)
    plt.show()

def dist_plot(col, data):
    f, ax = plt.subplots(figsize=(10, 5))
    sns.distplot(df[col], ax=ax)
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
    plt.show()

def bar_plot_ci(col, data):
    f, ax = plt.subplots(figsize=(10, 5))
    sns.barplot(x=col, y='target', data=df)
    plt.show()
```

```python
for col in obj_cols:
    bar_plot(col, df)
```
