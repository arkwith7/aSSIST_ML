> **Q 3. loc 함수를 이용해 25억 원을 초과하는 가격으로 거래된 아파트의 지역, 아파트명, 가격을 출력해 보세요.**

> **Q 4. '단가'라는 column을 만들고 '단가'를 내림차순으로 정리했을 경우, 이들 중 20억 원을 초과하는 아파트의 '지역', '아파트', '면적', '가격','단가' 데이터를 출력해 보세요.**

```python
apt['단가'] = apt['가격'] / apt['면적']
apt.head()
```

```python
filtered_price_df = apt.sort_values("단가", ascending=False)
filtered_price_df.head()
```

```python
filtered_price_df[filtered_price_df['가격'] > 20000, ['지역', '아파트', '면적', '가격','단가']]
```
