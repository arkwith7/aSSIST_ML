```python
# 새로운 data (column) 만들기
## 'size'라는 새로운 column을 만들어 'category'에 있는 값들을 'large'와 'small' 두 그룹으로 구분하기
### 'compact'와 'subcompact', '2seater'는 'small' 그룹으로 나머지 종류는 모두 'large' 그룹으로 구분할 것!!
mpg_data['category'].value_counts()
```

```python
# 새로운 data (column) 만들기
## 'size'라는 새로운 column을 만들어 'category'에 있는 값들을 'large'와 'small' 두 그룹으로 구분하기
### 'compact'와 'subcompact', '2seater'는 'small' 그룹으로 나머지 종류는 모두 'large' 그룹으로 구분할 것!!

mpg_data['size'] = np.where(((mpg_data['category'] == 'compact') | (mpg_data['category'] == 'subcompact') | (mpg_data['category'] == '2seater') ) , 'small', 'large')
mpg_data['size'].value_counts()
```

```python
# df.tail()
mpg_data.tail()
```

```python
# 'size'에 다시 한 가지 그룹을 더 추가해보기
## 'category'값 중 'midsize'는 'mid'그룹으로 구분하기, 결과적으로 모두 세 그룹이 되도록, 즉, 'small', 'mid', 'large' 그룹
mpg_data['size'] = np.where(mpg_data['category'] == 'midsize' , 'mid', mpg_data['size'])
mpg_data['size'].value_counts()
```

```python
# category가 'suv'인 애들을 대상으로 '전체 연비(total_mpg1)' column만들어 도심 연비와 고속도로 연비의 평균을 계산해서 넣어 주고 싶다면?
suv_data = mpg_data[mpg_data.category == 'suv']
print(suv_data.category.value_counts())
suv_data['total_mpg1'] = (mpg_data.cty + mpg_data.hwy) / 2
suv_data
```

```python
not_suv_data = mpg_data[mpg_data.category != 'suv']
print(not_suv_data.category.value_counts())
not_suv_data
```

```python
# category가 'suv'인 것과 아닌 다른 종류와 다시 합치기 --> pd.concat()
print("전체 데이터건수:",len(mpg_data))
print("SUV 데이터건수:",len(suv_data))
print("SUV이외 데이터건수:",len(not_suv_data))
df_con = pd.concat([not_suv_data, suv_data], axis=0, ignore_index=True)
df_con
```

```python
# category가 'suv'인 것과 다른 종류와 다시 합치기 --> pd.merge()
df_merge = pd.merge(suv_data, not_suv_data, how='outer')
df_merge
```
