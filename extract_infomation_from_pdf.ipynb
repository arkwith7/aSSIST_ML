```python
import os
import requests

# url = "https://github.com/arkwith7/aSSIST_ML/raw/main/asset/pdf/abl여30세.pdf"
url = "https://github.com/arkwith7/aSSIST_ML/raw/main/asset/pdf/無신여성시대건강보험.pdf"
output_folder = "pdf_files"

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

response = requests.get(url)
# filename = os.path.join(output_folder, "abl여30세.pdf")
filename = os.path.join(output_folder, "無신여성시대건강보험.pdf")

with open(filename, "wb") as f:
    f.write(response.content)

print(f"Downloaded file: {filename}")
```

```python
import pdfplumber

# keyword = "해약환급금 예시표"
# pdf_path = "pdf_files/abl여30세.pdf"
# keyword = "보장금액 예시표"
# pdf_path = "pdf_files/LKB.pdf"
keyword = "보험금의 종류 및 지급사유"
pdf_path = "pdf_files/無신여성시대건강보험.pdf"

with pdfplumber.open(pdf_path) as pdf:
    num_pages = len(pdf.pages)
    keyword_pages = []

    for i in range(num_pages):
        page = pdf.pages[i]
        text = page.extract_text()
        if keyword in text:
            keyword_pages.append(i)
            print("page : ",page)
            print("text : \n",text)

print(f"Keyword found in pages: {keyword_pages}")
```

```python
import tabula

output_csv_folder = "csv_files"
if not os.path.exists(output_csv_folder):
    os.makedirs(output_csv_folder)

for page_num in keyword_pages:
    # 테이블 추출
    tables = tabula.read_pdf(pdf_path, pages=page_num + 1, multiple_tables=True, pandas_options={'header': None})

    # CSV 파일로 저장
    for i, table in enumerate(tables):
        csv_filename = os.path.join(output_csv_folder, f"table_{page_num}_{i}.csv")
        table.to_csv(csv_filename, index=False)
        print(f"Saved table to: {csv_filename}")
```
